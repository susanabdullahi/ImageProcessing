```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/gaussian.ppm')  
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
filtered_image = cv2.GaussianBlur(image_rgb, (15, 15), 0) 
```

```python
plt.figure(figsize=(10, 5))
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')
```

```python
plt.subplot(1, 2, 2)
plt.imshow(filtered_image)
plt.title('Filtered Image')
plt.axis('off')
```

```python
plt.tight_layout()
plt.show()
```
