```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('images/sobel.jpeg')
```

```python
if len(image.shape) > 2:
    if image.shape[2] == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    elif image.shape[2] == 4:
        image = cv2.cvtColor(image, cv2.COLOR_BGRA2GRAY)
```

```python
if image.dtype != np.uint8:
    image = image.astype(np.uint8)
```

```python
equalized_image = cv2.equalizeHist(image)
```

```python
sobel_x_kernel = np.array([[-1, 0, 1],
                           [-2, 0, 2],
                           [-1, 0, 1]])

sobel_y_kernel = np.array([[-1, -2, -1],
                           [0, 0, 0],
                           [1, 2, 1]])
```

```python
sobel_x = cv2.filter2D(equalized_image, cv2.CV_64F, sobel_x_kernel)
sobel_y = cv2.filter2D(equalized_image, cv2.CV_64F, sobel_y_kernel)
```

```python
gradient_magnitude = np.sqrt(sobel_x ** 2 + sobel_y ** 2)
```

```python
threshold = 50 
edges = np.uint8(gradient_magnitude > threshold) * 255
```

```python
plt.figure(figsize=(18, 6))
```

```python
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
```

```python
plt.subplot(1, 3, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')
```

```python
plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title('Edges (Thresholded)')
plt.axis('off')
```

```python
plt.tight_layout()
plt.show()
```
