```python
import cv2
import numpy as np
```

```python
img = cv2.imread("images/segmentation.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
lower_green = np.array([0, 30, 0], dtype="uint8")
upper_green = np.array([100, 255, 100], dtype="uint8")

lower_blue = np.array([0, 0, 30], dtype="uint8")
upper_blue = np.array([100, 100, 255], dtype="uint8")

lower_red = np.array([30, 0, 0], dtype="uint8")
upper_red = np.array([255, 100, 100], dtype="uint8")
```

```python
mask_green = cv2.inRange(img, lower_green, upper_green)
mask_blue = cv2.inRange(img, lower_blue, upper_blue)
mask_red = cv2.inRange(img, lower_red, upper_red)
```

```python
mask_combined = cv2.bitwise_or(mask_green, cv2.bitwise_or(mask_blue, mask_red))
```

```python
segmented_image = cv2.bitwise_and(img_rgb, img_rgb, mask=mask_combined)
```
