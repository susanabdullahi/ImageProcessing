```python
import cv2
import numpy as np
```

```python
def gaussian_noise(image):
    row, col, ch = image.shape
    mean = 0
    var = 0.05
    sigma = var ** 0.5
  
    gauss = np.random.normal(mean, sigma, (row, col, ch))
    gauss = gauss.reshape(row, col, ch)
    # Adding the generated noise to the image
    noisy = image + gauss
    return noisy
```
