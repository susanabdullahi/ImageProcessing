```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def saltPepperNoise(image):
    row, col, ch = image.shape
    s_vs_p = 0.5
    amount = 0.04
    noisy = np.copy(image)
    
    # Salt noise
    num_salt = np.ceil(amount * image.size * s_vs_p)
    salt_coords = [np.random.randint(0, i, int(num_salt)) for i in image.shape]
    noisy[salt_coords[0], salt_coords[1], :] = 1
    
    # Pepper noise
    num_pepper = np.ceil(amount * image.size * (1.0 - s_vs_p))
    pepper_coords = [np.random.randint(0, i, int(num_pepper)) for i in image.shape]
    noisy[pepper_coords[0], pepper_coords[1], :] = 0
    
    return noisy
```

```python
img = cv2.imread("images/Bitplane.jpg") 

plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
img_norm = img.astype(np.float32) / 255.0
```

```python
noise_img = saltPepperNoise(img_norm)
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(noise_img, cmap='gray') 
plt.title('Image with Salt & Pepper Noise')
plt.axis('off')
plt.show()
```
