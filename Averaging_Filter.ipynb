```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = 'images/average.png' 
image = cv2.imread(image_path)
```

```python
if image is None:
    print("Could not read the image. Please check the file path.")
else:
   
    plt.figure(figsize=(8, 4))
```

```python
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
```

```python
 median_filtered = cv2.medianBlur(image, 3)
```

```python
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(median_filtered, cv2.COLOR_BGR2RGB))
plt.title('Median Filtered Image (3x3 kernel)')
plt.axis('off')
```

```python
plt.tight_layout()
plt.show()
```
