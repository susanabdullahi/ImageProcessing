```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/leplace.png') 
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
```

```python
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
```

```python
laplacian_kernel = np.array([[-1, -1, -1],
                             [-1,  8, -1],
                             [-1, -1, -1]])
```

```python
laplacian_custom = cv2.filter2D(blurred_image, cv2.CV_64F, laplacian_kernel)
```

```python
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
```

```python
plt.subplot(1, 3, 2)
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image (Gaussian Blur)')
plt.axis('off')
```

```python
plt.subplot(1, 3, 3)
plt.imshow(laplacian_custom, cmap='gray')
plt.title('Edge Enhanced (Custom Laplacian)')
plt.axis('off')
```

```python
plt.tight_layout()
plt.show()
```
